## Objectif du projet

## Revue de littérature

La classification automatique des déchets par vision par ordinateur est un domaine de recherche en plein essor, porté par les besoins croissants de tri sélectif et de recyclage intelligent.  
Les travaux existants se concentrent principalement sur deux axes :  
1. **L’utilisation de modèles de Deep Learning pour la reconnaissance d’images**,  
2. **L’adaptation de ces modèles à des contextes industriels réels**, où les conditions d’éclairage, de fond et de position des objets varient fortement.

---

### Jeux de données

Les premiers travaux se sont appuyés sur le dataset **TrashNet** (Thung & Yang, 2016), l’un des premiers jeux de données publics dédiés à la classification d’images de déchets.  
Cependant, ce jeu de données présente plusieurs limites : un nombre d’images restreint, des fonds uniformes et des conditions d’acquisition simplifiées (objets centrés, peu de bruit visuel).

Pour pallier ces limites, de nouveaux jeux de données plus riches et réalistes ont été proposés, notamment **TACO (Trash Annotations in Context)** et **RealWaste**, qui contiennent des images issues d’environnements réels avec des fonds variés, des éclairages non contrôlés et des objets parfois partiellement cachés ou déformés.  
Le dataset **RealWaste** est ainsi mieux adapté à l’évaluation de modèles destinés à une application industrielle.

---

### Modèles de classification

Les architectures de réseaux de neurones convolutionnels (**CNN**) comme **VGG**, **ResNet**, **EfficientNet** ou **MobileNet** ont montré une efficacité remarquable pour la classification d’images.  
Parmi elles, **MobileNetV2** est particulièrement adaptée aux environnements à ressources limitées grâce à sa structure compacte et optimisée.  
De plus, le **transfert d’apprentissage** (fine-tuning d’un modèle pré-entraîné sur ImageNet) permet d’obtenir de bonnes performances même avec un volume de données restreint, ce qui convient parfaitement au dataset RealWaste.

---

### Approches récentes et sobriété

Dans le cadre de l’**IA verte (Green AI)**, plusieurs études soulignent l’importance de réduire la consommation énergétique et la taille des modèles tout en maintenant des performances élevées.  
Les pratiques courantes incluent :
- l’utilisation de modèles légers comme **MobileNetV2** ou **EfficientNet-B0** plutôt que des architectures très profondes ;  
- la réduction du nombre d’époques d’entraînement grâce à un bon choix d’hyperparamètres et de techniques de régularisation ;  
- la mesure des compromis entre précision, temps de calcul et empreinte carbone.

---

### Limites et motivation du projet

Les approches existantes donnent souvent de bons résultats sur des images bien cadrées, mais leur généralisation à des environnements réels reste limitée.  
Notre projet vise donc à **étudier l’impact du fond visuel** sur les performances de classification, en comparant le dataset **RealWaste original** à une version **transformée sur fond noir**, simulant un environnement industriel (tapis roulant).  
Cette transformation a pour objectif de réduire le bruit de fond et d’améliorer la robustesse du modèle tout en maintenant un coût computationnel raisonnable.

### Séparation du dataset en jeux d’entraînement et de test

````python
# Dossiers principaux
base_dataset_dir = "dataset"                  # dossier où tu veux train/ et test/
raw_dataset_dir  = "dataset/RealWaste"        # dataset brut

# Classes (à partir du dataset brut uniquement)
classes = [
    c for c in sorted(os.listdir(raw_dataset_dir))
    if os.path.isdir(os.path.join(raw_dataset_dir, c))
]

# Construction du DataFrame df (filename + label)
data = []
for cls in classes:
    class_dir = os.path.join(raw_dataset_dir, cls)
    images = [f for f in os.listdir(class_dir) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
    for img_name in images:
        rel_path = os.path.join(cls, img_name)
        data.append({"filename": rel_path, "label": cls})

df = pd.DataFrame(data)

# Dossiers train/ et test/ à la racine de dataset/
train_dir = os.path.join(base_dataset_dir, "train")
test_dir  = os.path.join(base_dataset_dir, "test")

# Split stratifié
train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["label"],
    random_state=42
)

# Fonction de copie
def copy_images(dataframe, target_dir):
    for _, row in dataframe.iterrows():
        src = os.path.join(raw_dataset_dir, row['filename'])   # dataset/RealWaste/Classe/image.jpg
        dst_dir = os.path.join(target_dir, row['label'])       # dataset/train/Classe
        os.makedirs(dst_dir, exist_ok=True)
        dst = os.path.join(dst_dir, os.path.basename(row['filename']))
        if not os.path.exists(dst):
            shutil.copy2(src, dst)

print("Copying training images...")
copy_images(train_df, train_dir)

print("Copying test images...")
copy_images(test_df, test_dir)

print("Train/Test split completed!")

````

## 2. Prétraitement et Data Augmentation

Avant d’entraîner un modèle de Deep Learning, il faut **préparer les images** pour qu’elles soient toutes dans un format standard compris par le réseau.

Cette étape permet de :
- **redimensionner** les images à une taille fixe (ici 224×224) pour qu’elles aient toutes la même forme ;
- **normaliser** les valeurs des pixels entre 0 et 1 (au lieu de 0–255) pour stabiliser l’apprentissage ;
- **augmenter artificiellement** le dataset (rotations, zooms, flips horizontaux) afin d’améliorer la généralisation du modèle ;
- **diviser** le dataset en ensembles d’entraînement (80 %) et de validation (20 %) pour mesurer les performances sur des images jamais vues.

Ces opérations sont effectuées automatiquement à l’aide de la classe **`ImageDataGenerator`** de Keras.

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Dossiers (doivent correspondre à ce que tu as fait dans le split)
base_dataset_dir = "dataset"
train_dir = os.path.join(base_dataset_dir, "train")  # dataset/train/Classe/...
test_dir  = os.path.join(base_dataset_dir, "test")   # dataset/test/Classe/...

IMG_SIZE   = (224, 224)
BATCH_SIZE = 32

# Data augmentation + normalisation pour le CNN simple
cnn_train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2    # 20% de train_dir pour la validation
)

# Générateur TRAIN (environ 64% des données)
cnn_train_generator = cnn_train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

# Générateur VAL (environ 16% des données)
cnn_val_generator = cnn_train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# Générateur TEST (les 20% mis de côté, sans augmentation)
cnn_test_datagen = ImageDataGenerator(rescale=1./255)

cnn_test_generator = cnn_test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

num_classes   = cnn_train_generator.num_classes
class_indices = cnn_train_generator.class_indices
print("Classes (indices) :", class_indices)
````

### Interprétation des résultats du prétraitement

L’ensemble a été automatiquement divisé en :
- **3 803 images** pour l’apprentissage (80 %)  
- **949 images** pour la validation (20 %)

Les classes détectées correspondent bien aux catégories du dataset RealWaste :  
*Cardboard, Food Organics, Glass, Metal, Miscellaneous Trash, Paper, Plastic, Textile Trash, Vegetation.*

Chaque classe s’est vu attribuer un **indice numérique** (de 0 à 8), utilisé par le modèle pendant l’entraînement.  
Cette étape confirme que la structure du dataset est correcte et que les images sont prêtes pour la phase d’apprentissage.

### Visualisation des images après augmentation

Avant de passer à la phase d’entraînement, il est important de vérifier visuellement le résultat du prétraitement.  
Les générateurs Keras appliquent automatiquement des transformations aléatoires (rotations, zooms, décalages, flips) sur les images d’entraînement afin de **simuler une plus grande diversité de données**.

La cellule suivante permet d’afficher un échantillon d’images **après augmentation**, c’est-à-dire telles qu’elles sont réellement vues par le modèle pendant l’apprentissage.

````python
images, labels = next(cnn_train_generator)

plt.figure(figsize=(12, 8))
for i in range(12):
    plt.subplot(3, 4, i+1)
    plt.imshow(images[i])
    cls_idx = np.argmax(labels[i])
    cls_name = list(class_indices.keys())[list(class_indices.values()).index(cls_idx)]
    plt.title(cls_name, fontsize=8)
    plt.axis("off")

plt.suptitle("Exemples d'images après data augmentation (CNN simple)", fontsize=14)
plt.tight_layout()
plt.show()
````

## 3. Modèle de base (CNN simple)

**But.** Construire une baseline légère et propre pour la classification RealWaste.  
Cette baseline servira de point de comparaison avec un modèle pré-entraîné (MobileNetV2).

**Idée.** Un petit CNN en 3 blocs convolution → batch norm → max-pooling, puis une tête dense simple.  
On compile avec `categorical_crossentropy` et on suit l’`accuracy`.

**Entrées/sorties.**  
- Entrée : images 224×224×3 (issues du générateur avec normalisation).  
- Sortie : vecteur de probas sur 9 classes (softmax).


### 3.1 Construction et compilation du modèle

On définit une architecture minimale, robuste et lisible :  
- 3 blocs Conv2D (32 → 64 → 128), `ReLU`, `BatchNormalization`, `MaxPooling2D` ;  
- `Flatten` + `Dropout` + `Dense(128)` ;  
- `Dense(9, softmax)` pour la classification multi-classes.

On compile avec **Adam (1e-3)** et la **loss catégorielle**.  
Objectif : obtenir une baseline stable avant tout réglage fin.


````python
# Construction du modèle CNN simple
def build_basic_cnn(input_shape=(224, 224, 3), num_classes=9):
    model = models.Sequential([
        layers.Input(shape=input_shape),

        layers.Conv2D(32, (3, 3), activation="relu", padding="same"),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),

        layers.Conv2D(64, (3, 3), activation="relu", padding="same"),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),

        layers.Conv2D(128, (3, 3), activation="relu", padding="same"),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),

        layers.Flatten(),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax")
    ])
    return model

cnn_model = build_basic_cnn(input_shape=(224, 224, 3), num_classes=num_classes)
cnn_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

cnn_model.summary()
````

### 3.2 Entraînement : callbacks et protocole

**Callbacks utilisés :**  
- **EarlyStopping** *(patience=5, monitor=val_loss)* : évite le surapprentissage et réduit le coût.  
- **ReduceLROnPlateau** : diminue le learning rate si la validation stagne.  
- **ModelCheckpoint** : sauvegarde automatique du meilleur modèle.

**Protocole.**  
- Entraînement sur l’ensemble *train*, évaluation continue sur *validation*.  
- Nombre d’époques nominal : **20** (interrompu plus tôt si EarlyStopping).


````python
# Entraînement du CNN simple
EPOCHS = 10

history_cnn = cnn_model.fit(
    cnn_train_generator,
    validation_data=cnn_val_generator,
    epochs=EPOCHS
)
````

Au fil des époques, la **loss** diminue progressivement (de 9 à environ 2.09), ce qui montre que le modèle apprend.  
Cependant, l’**accuracy** reste faible, autour de **0.20 - 0.24** en apprentissage et **0.10 - 0.20** en validation.  
Le modèle converge lentement et ne parvient pas encore à bien généraliser : c’est un comportement typique d’un **CNN entraîné from scratch** sur un dataset complexe comme RealWaste.

### 3.3 Courbes d’apprentissage

On trace les **courbes loss** (train/val) et **accuracy** (train/val).  
Objectif : vérifier la stabilité de l’apprentissage, la convergence et détecter un éventuel écart train/val (surapprentissage).

````python
# Fonction utilitaire pour tracer les courbes d'apprentissage
def plot_history(history, title_prefix=""):
    acc      = history.history["accuracy"]
    val_acc  = history.history["val_accuracy"]
    loss     = history.history["loss"]
    val_loss = history.history["val_loss"]

    epochs_range = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label="Train acc")
    plt.plot(epochs_range, val_acc, label="Val acc")
    plt.title(f"{title_prefix}Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label="Train loss")
    plt.plot(epochs_range, val_loss, label="Val loss")
    plt.title(f"{title_prefix}Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_history(history_cnn, "CNN simple – ")
````

Les courbes montrent une **forte baisse de la loss** dès les premières époques, signe que le modèle apprend les motifs de base.  
Cependant, l’**accuracy** reste faible (≈ 20–30 %) et assez instable entre train et validation.  
Cette variabilité indique un **apprentissage limité** : le modèle parvient à s’adapter partiellement aux données d’entraînement mais **ne généralise pas bien** sur la validation.  
Cela confirme que le **CNN simple** atteint rapidement ses limites sur un dataset complexe comme RealWaste.

### 3.4 Évaluation sur la validation

On reporte la **loss** et l’**accuracy** sur l’ensemble de validation.  
Cette mesure sert de **référence** avant :  
1) l’utilisation de **MobileNetV2** (transfert d’apprentissage),  
2) l’évaluation avancée (F1, matrice de confusion),  
3) l’analyse **Grad-CAM**.

````python
from sklearn.metrics import classification_report, confusion_matrix, f1_score

# Prédictions sur le set de test
cnn_test_generator.reset()  # pour être sûr de parcourir tout le test correctement
y_prob_cnn = cnn_model.predict(cnn_test_generator)
y_pred_cnn = np.argmax(y_prob_cnn, axis=1)

# Vrais labels
y_true_cnn = cnn_test_generator.classes              # indices numériques
class_names = list(cnn_test_generator.class_indices.keys())  # noms de classes dans l'ordre

# Metrics globales
acc_cnn = np.mean(y_pred_cnn == y_true_cnn)
f1_macro_cnn = f1_score(y_true_cnn, y_pred_cnn, average="macro")

print(f"Accuracy (test) CNN simple : {acc_cnn:.4f}")
print(f"F1-macro (test) CNN simple : {f1_macro_cnn:.4f}\n")

# Rapport de classification détaillé
print("Classification report (CNN simple):\n")
print(classification_report(y_true_cnn, y_pred_cnn, target_names=class_names))

# Matrice de confusion
cm_cnn = confusion_matrix(y_true_cnn, y_pred_cnn)

plt.figure(figsize=(8, 6))
plt.imshow(cm_cnn, interpolation="nearest")
plt.title("Matrice de confusion – CNN simple")
plt.colorbar()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45, ha="right")
plt.yticks(tick_marks, class_names)
plt.xlabel("Prédit")
plt.ylabel("Réel")
plt.tight_layout()
plt.show()
````

Le modèle atteint une **accuracy de 34,7 %** sur l’ensemble de validation, avec une **loss d’environ 1.86**.  
Ces résultats sont modestes mais cohérents pour un **CNN entraîné from scratch** sur un dataset complexe.  
Le réseau parvient à apprendre certaines caractéristiques visuelles, mais sa capacité de généralisation reste limitée.  
Cette mesure servira donc de **référence (baseline)** avant d’introduire le **transfert d’apprentissage** avec MobileNetV2.

## 4. MobileNetV2 (transfert d’apprentissage)

Objectif : réutiliser un modèle **pré-entraîné sur ImageNet** (MobileNetV2) et **n’entraîner que la tête de classification**.  
Avantages : meilleure précision avec peu d’époques et un coût réduit.

### 4.1 — Générateurs Keras avec le prétraitement MobileNetV2

On utilise `preprocess_input` (MobileNetV2) **dans les générateurs**.  
Du coup, **pas de Rescaling** dans le modèle (sinon double normalisation).

````python
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Générateurs avec preprocess_input (spécifique à MobileNetV2)
mnet_train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2   # on recrée un train/val à partir de train_dir
)

# Entraînement (64 %)
mnet_train_generator = mnet_train_datagen.flow_from_directory(
    train_dir,                 # ex: "dataset/train"
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

# Validation (16 %)
mnet_val_generator = mnet_train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# Test (20 %) – pas de data augmentation, mais même preprocess_input
mnet_test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

mnet_test_generator = mnet_test_datagen.flow_from_directory(
    test_dir,                  # ex: "dataset/test"
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

num_classes_mnet   = mnet_train_generator.num_classes
class_indices_mnet = mnet_train_generator.class_indices
print("Classes (indices) – MobileNetV2 :", class_indices_mnet)
````

### 4.2 — Construction du modèle (feature extractor)

- `include_top=False` pour enlever la tête ImageNet d’origine,  
- **gel** des poids MobileNetV2,  
- tête simple : `GlobalAveragePooling2D → Dense(128, ReLU) → Dropout → Dense(num_classes, softmax)`.

````python
from tensorflow.keras import layers, models
import tensorflow as tf

# Base MobileNetV2 pré-entraînée sur ImageNet, sans la top layer
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
)

# On gèle les poids de la base pour un premier entraînement
base_model.trainable = False

# Construction de la tête de classification
inputs = layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes_mnet, activation='softmax')(x)

mnet_model = models.Model(inputs, outputs)
mnet_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

mnet_model.summary()
````

### 4.3 — Entraînement 

````python
EPOCHS_MNET = 15

history_mnet = mnet_model.fit(
    mnet_train_generator,
    validation_data=mnet_val_generator,
    epochs=EPOCHS_MNET
)
````

### 4.4 léger fine-tuning

````python
# On dégèle une partie de la base pour un léger fine-tuning
base_model.trainable = True

# On choisit un certain nombre de couches à laisser gelées
fine_tune_at = 100  # par exemple : on gèle les 100 premières couches

for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

mnet_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),  # plus petit LR
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

EPOCHS_FINE = 5

history_mnet_ft = mnet_model.fit(
    mnet_train_generator,
    validation_data=mnet_val_generator,
    epochs=EPOCHS_FINE
)
````

### 4.5 Évaluation de MobileNetV2 sur le test set

On trace les courbes **loss/accuracy** et on mesure l’accuracy sur la **validation**.  
Ce résultat doit être **nettement supérieur** à la baseline CNN.

````python
# Évaluation finale de MobileNetV2 fine-tuné sur le test set

mnet_test_generator.reset()
y_prob_mnet_ft = mnet_model.predict(mnet_test_generator)
y_pred_mnet_ft = np.argmax(y_prob_mnet_ft, axis=1)

y_true_mnet = mnet_test_generator.classes
class_names_mnet = list(mnet_test_generator.class_indices.keys())

acc_mnet_ft = np.mean(y_pred_mnet_ft == y_true_mnet)
f1_macro_mnet_ft = f1_score(y_true_mnet, y_pred_mnet_ft, average="macro")

print(f"Accuracy (test) MobileNetV2 fine-tuné : {acc_mnet_ft:.4f}")
print(f"F1-macro (test) MobileNetV2 fine-tuné : {f1_macro_mnet_ft:.4f}\n")

print("Classification report (MobileNetV2 fine-tuné):\n")
print(classification_report(y_true_mnet, y_pred_mnet_ft, target_names=class_names_mnet))

# ===== Matrice de confusion avec les mêmes couleurs/valeurs que ton pote =====
cm_mnet_ft = confusion_matrix(y_true_mnet, y_pred_mnet_ft)

plt.figure(figsize=(10, 8))
sns.heatmap(
    cm_mnet_ft,
    annot=True,            # affiche les valeurs dans les cases
    fmt='d',               # valeurs entières
    cmap='Blues',          # même colormap
    xticklabels=class_names_mnet,
    yticklabels=class_names_mnet
)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix – MobileNetV2 (fine-tuning)')
plt.tight_layout()
plt.show()
````
